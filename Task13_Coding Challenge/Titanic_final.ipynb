# Question3:Pivot Table

### a. Sex - age

```python
gender_age = pd.pivot_table(df, index = ["Sex","Age"], aggfunc=np.sum)
gender_age.head(10)
```

### b. Sex – Passenger_class

```python
gender_pclass = pd.pivot_table(df, index = ["Sex","Pclass"], aggfunc=np.sum)
gender_pclass.head()
```

### c. Sex – survived

```python
gender_survived = pd.pivot_table(df, index = ["Sex","Survived"], aggfunc=np.sum)
gender_survived.head()
```

# Question5

### a. adult male, adult female[Male = True & Females = False]

```python
person_age = df["Age"] > 18
adult_male = df["Sex"] == "male"

pd.pivot_table(df, index = [person_age & adult_male], aggfunc=np.sum)
```

### b. Male and female children

```python
childs_age = df["Age"] < 18
childs_male = df["Sex"] == "male"
pd.pivot_table(df, index = [childs_age & childs_male], aggfunc=np.sum)
```

### c. Children are any group < =18 age[True = total numbers of childern]

```python
under_age = df["Age"] <= 18
pd.pivot_table(df, index = under_age, aggfunc=np.sum)
```
